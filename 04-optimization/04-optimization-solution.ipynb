### 1. Loss Functions and Gradient Descent

**Task:** Understand how loss functions guide optimization and implement gradient descent with backpropagation.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Example data
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])

# Define a simple linear model
model = nn.Linear(1, 1)

# Loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
epochs = 50
for epoch in range(epochs):
    optimizer.zero_grad()
    predictions = model(x)
    loss = criterion(predictions, y)
    loss.backward()  # Compute gradients
    optimizer.step()  # Update weights

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item()}")

print("Trained Model Parameters:")
print("Weight:", model.weight.item())
print("Bias:", model.bias.item())
```

### 2. Experiment with Learning Rates

**Task:** Modify the learning rate to observe its effect on the optimization process.

```python
# Experiment with different learning rates
learning_rates = [0.1, 0.01, 0.001]
for lr in learning_rates:
    model = nn.Linear(1, 1)
    optimizer = optim.SGD(model.parameters(), lr=lr)

    print(f"Training with Learning Rate = {lr}")
    for epoch in range(20):
        optimizer.zero_grad()
        predictions = model(x)
        loss = criterion(predictions, y)
        loss.backward()
        optimizer.step()

        if epoch % 5 == 0:
            print(f"Epoch {epoch}: Loss = {loss.item()}")
    print("-" * 30)
```

**Exercise 1: Implement Gradient Clipping**
1. Modify the training loop to clip the gradients at a maximum norm of `1.0`.
2. Observe the effect of gradient clipping on training stability.

**Solution:**

```python
# Solution for Exercise 1
# Gradient Clipping in Training Loop
model = nn.Linear(1, 1)  # Simple linear model
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(50):
    optimizer.zero_grad()
    predictions = model(x)
    loss = criterion(predictions, y)
    loss.backward()

    # Clip gradients
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item()}")
```

### 3. Regularization with Dropout

**Task:** Introduce dropout as a technique to prevent overfitting.

```python
# Define a model with Dropout
class DropoutExample(nn.Module):
    def __init__(self):
        super(DropoutExample, self).__init__()
        self.fc1 = nn.Linear(10, 10)
        self.dropout = nn.Dropout(p=0.5)
        self.fc2 = nn.Linear(10, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Initialize the model
model = DropoutExample()
print(model)

# Example forward pass
example_input = torch.randn(1, 10)
output = model(example_input)
print("Output with Dropout:", output)
```

**Exercise 1: Modify Dropout Rate**
Change the dropout rate to `0.3` and `0.7` and observe its impact during forward passes.

**Solution:**

```python
# Change dropout rate
for p in [0.3, 0.7]:
    model = DropoutExample()
    model.dropout = nn.Dropout(p=p)
    print(f"Dropout rate = {p}")

    output = model(example_input)
    print("Output with Dropout:", output)
    print("-" * 30)
```

**Exercise 2: Compare Optimizers**
1. Train the model using `SGD`, `Adam`, and `RMSprop` optimizers.
2. Compare the final training loss for each optimizer.

**Solution:**

```python
# Solution for Exercise 2
optimizers = {
    "SGD": optim.SGD(model.parameters(), lr=0.01),
    "Adam": optim.Adam(model.parameters(), lr=0.01),
    "RMSprop": optim.RMSprop(model.parameters(), lr=0.01)
}

for name, opt in optimizers.items():
    model = nn.Linear(1, 1)  # Reinitialize model for each optimizer
    print(f"Training with {name} optimizer")
    for epoch in range(20):
        opt.zero_grad()
        predictions = model(x)
        loss = criterion(predictions, y)
        loss.backward()
        opt.step()

        if epoch % 5 == 0:
            print(f"Epoch {epoch}: Loss = {loss.item()}")
    print("-" * 30)
```

#### Exercise 1: Custom Loss Function
Create a custom loss function that computes the absolute error instead of the mean squared error. Use it to train the linear model and compare results with MSELoss.

**Solution:**

```python
# Custom Absolute Error Loss
class AbsoluteErrorLoss(nn.Module):
    def __init__(self):
        super(AbsoluteErrorLoss, self).__init__()

    def forward(self, predictions, targets):
        return torch.mean(torch.abs(predictions - targets))

# Training with the custom loss
absolute_loss = AbsoluteErrorLoss()
model = nn.Linear(1, 1)
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(50):
    optimizer.zero_grad()
    predictions = model(x)
    loss = absolute_loss(predictions, y)
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Absolute Error Loss = {loss.item()}")
```

**Exercise 3: L2 Regularization (Weight Decay)**
1. Add L2 regularization to the `SGD` optimizer with a weight decay factor of `0.01`.
2. Compare the training loss with and without weight decay.

**Solution:**

```python
# Solution for Exercise 3
# Training with and without L2 Regularization
for weight_decay in [0.0, 0.01]:
    print(f"Training with Weight Decay = {weight_decay}")
    model = nn.Linear(1, 1)
    optimizer = optim.SGD(model.parameters(), lr=0.01, weight_decay=weight_decay)

    for epoch in range(20):
        optimizer.zero_grad()
        predictions = model(x)
        loss = criterion(predictions, y)
        loss.backward()
        optimizer.step()

        if epoch % 5 == 0:
            print(f"Epoch {epoch}: Loss = {loss.item()}")
    print("-" * 30)
```

#### Exercise 2: Implement Momentum in Gradient Descent
Modify the optimizer to use SGD with momentum. Experiment with momentum values of `0.5`, `0.9`, and `0.99`. Observe its effect on training speed.

**Solution:**

#### Exercise 3: Dropout Effect on Model Performance
Train a simple neural network with and without dropout. Compare the final training loss in both cases.

**Solution:**

```python
# Model with and without Dropout
dropout_rates = [0.0, 0.5]
for rate in dropout_rates:
    class SimpleDropoutModel(nn.Module):
        def __init__(self, dropout_rate):
            super(SimpleDropoutModel, self).__init__()
            self.fc1 = nn.Linear(10, 10)
            self.dropout = nn.Dropout(p=dropout_rate)
            self.fc2 = nn.Linear(10, 1)

        def forward(self, x):
            x = torch.relu(self.fc1(x))
            x = self.dropout(x)
            x = self.fc2(x)
            return x

    # Train the model
    model = SimpleDropoutModel(dropout_rate=rate)
    optimizer = optim.SGD(model.parameters(), lr=0.01)
    example_input = torch.randn(100, 10)
    example_target = torch.randn(100, 1)

    for epoch in range(20):
        optimizer.zero_grad()
        predictions = model(example_input)
        loss = criterion(predictions, example_target)
        loss.backward()
        optimizer.step()

        if epoch % 5 == 0:
            print(f"Dropout Rate {rate}, Epoch {epoch}: Loss = {loss.item()}")
    print("-" * 30)
```
